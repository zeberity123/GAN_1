```python
n = 10
import numpy as np
result = np.zeros((n,n))
result
```

```python
num = 1
i = 0
j = -1
trans_num = 1

while n > 0 :
# 열 증가
    for _ in range(n):
        j += trans_num
        result[i][j] = num
        num += 1
    n -= 1
    #행 증가
    for _ in range(n):
        i += trans_num
        result[i][j] = num
        num += 1
    trans_num *= -1
result
```
